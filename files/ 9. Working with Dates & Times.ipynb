# Working With Dates Exercise

```python
import pandas as pd
```

```python
# Import the "kc_house_data.csv" and parse the "date" column in a single line

houses = pd.read_csv('data/kc_house_data.csv', parse_dates = ['date'])
```

```python
# Calculate how many days the dataset spans

result = houses['date'].max() - houses['date'].min()
result.days
```

```python
#find the date that had the most house sales
#find all the homes sold on that date

# Variant 1

houses['date'].value_counts().head(1)
houses[(houses['date'] == '2014-06-23')]
```

```python
# Variant 2

top_date = houses['date'].mode(1)[0]
houses[houses['date'] == top_date]
```

```python
# Create a pie chart showing the years that home sales took place in (2014 and 2015)

houses.date.dt.year.value_counts().plot(kind = 'pie')
```

```python
#Find all the homes sold in 2014, sorted from earliest to latest

houses[houses['date'].dt.year == 2014].sort_values('date', ascending = True)
```

```python
#Count up the number of sales that took place in each month of the year.  Which month had the most?

houses['month'] = houses['date'].dt.month
houses['month'].value_counts()
```

```python
# Find the homes sold between May 1st 2014 and May 1st 2015
# Create a bar plot showing the total number of sales per month in that period
# The x axis should be in calendar order (1-12)

year_sales = houses[houses['date'].between('2014-05-01', '2015-05-01')].sort_values('date')
year_sales['date'].dt.month.value_counts().sort_index(ascending = True).plot(kind = 'bar')
```

```python
# Count up the number of homesales by day of the week

houses['date'].dt.dayofweek.value_counts()
```

```python
# Create a line plot showing the total number of sales by week of the year number (1-52)

houses['date'].dt.isocalendar().week.value_counts().sort_index().plot(kind = 'line')
```

```python
#Create a bar plot showing the total number of sales that took place in December, January, and February

houses['date'].dt.month.value_counts().loc[[1,2,12]].plot(kind = 'bar')
```

```python
#Work with the homes sold between May 1st 2014 and May 1st 2015
#Within that year, find the waterfront homes that were sold
#Which quarter of that year had the most waterfront home sales?  The least?
#Create a bar plot showing the number of waterfront home sales per quarter


dates = houses[houses['date'].between('2014-05-01', '2015-05-01')]
waterfront_houses = dates[dates['waterfront'] == 1]
waterfront_houses['date'].dt.quarter.value_counts().sort_index().plot(kind = 'bar')
```

## More Useful Info:

```python
# to_datetime method converts 'messy' dates to yy-mm-dd format

pd.to_datetime('Dec. 6th 2018')
```

```python
pd.to_datetime('December 6th 2018')
```

```python
pd.to_datetime('2018.12.06')
```

```python
pd.to_datetime('2018/12/06')
```

```python
pd.to_datetime('2018/12/06 4pm')
```

```python
# when it is not clear what numbers should be detected as a year, 
# one may use yearfirst argument

pd.to_datetime('18/12/06', yearfirst = True)
```

```python
# alternatively it is possible to use dayfirst argument

pd.to_datetime('18/12/06', dayfirst = True)
```

```python
# it is also possible to set specific date format: 
# https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior

pd.to_datetime('18/12/06', format = '%y/%m/%d')
```
