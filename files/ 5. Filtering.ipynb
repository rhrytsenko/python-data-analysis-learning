# Filtering Data Exercise
Answer the following questions using the `bestsellers.csv` dataset

```python
import pandas as pd
bestsellers = pd.read_csv('data/bestsellers.csv')
```

```python
# Find the books written by Pete Souza

bestsellers[bestsellers['Author'] == 'Pete Souza']
```

```python
# Find the books that are under 10 dollars

bestsellers[bestsellers['Price'] < 10]
```

```python
# Find the books that have a price between 50 and 60 dollars

bestsellers[(bestsellers['Price'] > 50) & (bestsellers['Price'] < 60)]
```

```python
# Find the books that have a price between 50 and 60 dollars
# (variables method)

above_50 = bestsellers['Price'] > 50
under_60 = bestsellers['Price'] < 60

bestsellers[above_50 & under_60]
```

```python
# Find the books that have a price between 50 and 60 dollars
# (through 'between' function)

bestsellers[bestsellers['Price'].between(50,60)]
```

```python
# Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens

bestsellers[bestsellers['Author'].isin(['Kristin Hannah', 'Andy Weir', 'Delia Owens'])]
```

```python
# Find the Non Fiction books that are rated 4.9

bestsellers[(bestsellers['Genre'] == 'Non Fiction') & (bestsellers['User Rating'] == 4.9)]
```

```python
# Find the fiction book with the lowest User Rating

bestsellers[bestsellers['Genre'] == 'Fiction'].sort_values('User Rating').head(1)
```

```python
# Find 2012's top 5 Fiction books with the most Reviews

bestsellers[(bestsellers['Year'] == 2012) & 
            (bestsellers['Genre'] == 'Fiction')].sort_values('Reviews', ascending = False).head(5)
```

```python
# Create a bar plot showing the 5 authors who have the most books with a rating under 4.5

f_table = bestsellers[bestsellers['User Rating'] < 4.5]
f_table['Author'].value_counts().head(5).sort_values(ascending = True).plot(kind = 'bar')
```

```python
# Bitwise Negation "~"

titanic = pd.read_csv('data/titanic.csv')

women = titanic.sex == 'female'
titanic[~women].head(6)

# No need to change all the logical operators if you have complex query
```

```python
# Operator OR is typed as "|"

houses = pd.read_csv('data/kc_house_data.csv')

houses[(houses['yr_built'] >= 2014) | (houses['yr_renovated'] >= 2014)]
```
